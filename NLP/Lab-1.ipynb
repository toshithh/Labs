Sentences: "Hello how are you?"

2) StopWords - Remove stopwords such as articles, etc.
They do not provide any significant value to the sentence hence,
we need to remove them.

3) Root word extraction - Stemming

```python
import nltk
nltk.download("punkt")
nltk.download("wordnet")
nltk.download("stopwords")
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer

def tokenize(text):
  tokens = word_tokenize(text)
  stop_words = set(stopwords.words("english"))
  tokens = [token for token in tokens if token.lower() not in stop_words]

  stemmer = PorterStemmer()
  stemmed_tokens = [stemmer.stem(token) for token in tokens]

  lemmatizer = WordNetLemmatizer()
  lemmatized = [lemmatizer.lemmatize(token) for token in tokens]

  return tokens, stemmed_tokens, lemmatized


text = "Hello friends! Finland's Chai pilo while dancing drank m.p.h. show-time"

print(*tokenize(text), end="\n", sep="\n")

##### Text File
file = open("/content/drive/MyDrive/Colab Notebooks/NLP/semaphore_text.txt", 'r')
text = file.read()
file.close()

print("\n\n", *tokenize(text), end="\n", sep="\n")
```
