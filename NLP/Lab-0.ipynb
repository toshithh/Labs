```python
import re

file = open("/content/drive/MyDrive/Colab Notebooks/NLP/semaphore_text.txt", "r")
text = file.read()
file.close()
print(text)

comma_sep = text.split(",")
comma = []
for i in range(len(comma_sep)-1):
  comma.append((re.findall("[a-zA-Z0-9]+", comma_sep[i])[-1], re.findall("[a-zA-Z0-9]+", comma_sep[i])[0]))
print(comma)

unique = re.findall("[a-zA-Z0-9]+", text.lower())
unique_words = []
for i in range(len(unique)):
  if unique[i] in unique[:i]+unique[i+1:]:
    continue
  else:
    unique_words.append(unique[i])

print(unique_words)
print(len(unique_words), len(set(unique_words)))


print(len(re.findall("[a-zA-Z0-9]+", text)))
```
