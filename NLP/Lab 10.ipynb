```python
import re
from collections import Counter

file = open("/content/drive/MyDrive/Colab Notebooks/NLP/snowy_story.txt", "r")
text = file.read()
file.close()


def preprocess_text(text):
  text = text.lower()
  text = re.sub(r'[^a-zA-Z0-9\s]', '', text)  #remove non-alphanumeric characters
  text = re.sub(r'\s+', ' ', text).strip()  #remove extra white space
  return text

def tokenize(text):
  return text.split()

def calc_tf(tokens):
  tf = dict()
  for x in tokens:
    if x in list(tf.keys()):
      tf[x] += 1
    else:
      tf[x] = 1
  for x in list(tf.keys()):
    tf[x] = tf[x]/len(tokens)
  return tf

def calc_idf(docs):
  idf = {}
  tot_docs = len(docs)
  all_tokens = [token for doc in docs for token in set(doc)]
  token_counter = Counter(all_tokens)
  for token, count in token_counter.items():
    idf[token] = tot_docs/count
  return idf

def calc_tfidf(tf, idf):
  tfidf = {x: tf[x]*idf[x] for x in tf.keys()}
  return tfidf

def generate_summary(text, ratio=0.2):
  text = preprocess_text(text)
  tokens = tokenize(text)
  tf = calc_tf(tokens)
  idf = calc_idf([tokens])
  tfidf = calc_tfidf(tf, idf)
  sorted_tokens = sorted(tfidf.items(), key=lambda x: x[1], reverse=True)
  num_tokens = int(len(tokens)*ratio)
  summary_tokens = [token for token, _ in sorted_tokens[:num_tokens]]
  summary = ' '.join(summary_tokens)
  return summary


gen = generate_summary(text, 0.2)
i = 0
for x in gen.split(" "):
  if i == 15:
    print(x)
    i = 0
  else:
    print(x, end=" ")
    i+=1
```
